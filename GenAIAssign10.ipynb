```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.to("cuda")
```

```python
def generate_image(prompt, steps=50, guidance_scale=7.5):
    image = pipe(prompt, num_inference_steps=steps, guidance_scale=guidance_scale).images[0]
    return image
```

```python
prompt = "A futuristic city with neon lights and flying cars at night"
image = generate_image(prompt)
image.show()
image.save("generated_image.png")
```
